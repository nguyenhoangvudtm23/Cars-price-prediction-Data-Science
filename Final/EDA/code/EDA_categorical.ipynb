```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('D:\Study\PythonPrj\pythonProject\Data_Science\Cars\cars_data_cleaning.csv')
df = df.drop(['url', 'weight', 'trailerWeight', 'name', 'model', 'numberOfAxles'], axis=1)
df = df[df['price'].notna()]
df.head()
```

```python
#skewness and kurtosis
print("Skewness: %f" % df['price'].skew())
print("Kurtosis: %f" % df['price'].kurt())
```

```python
# relationship with numerical variables
def relation(var):
    data = pd.concat([df['price'], df[var]], axis=1)
    data.plot.scatter(x=var, y='price')
```

```python
def categorical_relation(var):
    data = pd.concat([df['price'], df[var]], axis=1)
    f, ax = plt.subplots(figsize=(15, 6))
    fig = sns.boxplot(x=var, y="price", data=data)
    return fig
```

```python
categorical_relation('fuelType')
```
