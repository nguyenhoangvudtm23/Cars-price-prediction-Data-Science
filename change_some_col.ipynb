```python
# Find some error in emissionsCO2
d = dict()
for value in car_data["emissionsCO2"]:
    number = value.split(" ")[0]
    try:
        int(number)
    except:
        if number not in d:
            d[number] = 1
        else:
            d[number] += 1
print(d)
```

```python
# Gasoline / bio ethanol is same as Bio ethanol / gasoline
fuel_class = ["Gasoline", "Diesel", "Bio ethanol", "Natural gas / gasoline", "LPG / gasoline", "Natural gas", "LPG", "Bio ethanol / gasoline", "Gasoline / bio ethanol"]
num_class = len(fuel_class) - 1 + 1 # minus 1 trung + 1 null

repair = []
for value in car_data['fuelType']:
    if value in fuel_class:
        if value == "Bio ethanol / gasoline" or value == "Gasoline / bio ethanol":
            repair.append(len(fuel_class) - 2)
        else:
            repair.append(fuel_class.index(value))
    else:
        repair.append(len(fuel_class) - 1)
print(repair[:10])
print( car_data['fuelType'].iloc[:10])
```

```python
car_data['fuelType'] = pd.Series(repair)
```
